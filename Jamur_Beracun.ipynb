```python
label = LabelEncoder()
labels = {}
for col in df.columns:
    label.fit(df[col])
    labels[col] = label.transform(df[col])
    labels
print(labels['class'])
```

```python
df_labels = pd.DataFrame(labels)
df_labels
```

```python
# split train/test data set
xtr, xts, ytr, yts = train_test_split(
    df_labels.drop("class", axis = 1), 
    df["class"], 
    test_size=.2
)
len(xtr), len(xts), len(ytr), len(yts)
```
